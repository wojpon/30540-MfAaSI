```python
import numpy as np

from skimage import io
from skimage.util import img_as_ubyte, img_as_float

from skimage import exposure
```

```python
img_before = io.imread("/Users/wojpon/Developer/DTU/MfAnSI/remote-sensing/coral-reef-project/image1988.tif")
img_after = io.imread("/Users/wojpon/Developer/DTU/MfAnSI/remote-sensing/coral-reef-project/image2011.tif")
```

```python
img_before = img_as_float(exposure.rescale_intensity(img_before))
img_after = img_as_float(exposure.rescale_intensity(img_after))
```

# Image Before RGB & FCC

```python
img_before_R = img_before[:, :, 2]
img_before_G = img_before[:, :, 1]
img_before_B = img_before[:, :, 0]
img_before_N = img_before[:, :, 3]
```

```python
img_before_RGB = np.stack([img_before_R, img_before_G, img_before_B], axis=2)
```

```python
io.imshow(img_before_RGB)
```

```python
# io.imsave("outputs/img_before_RGB.png", img_before_RGB)
```

```python
img_before_NIR = np.stack([img_before_R, img_before_G, img_before_N], axis=2)
```

```python
io.imshow(img_before_NIR)
```

```python
# io.imsave("outputs/img_before_FCC.png", img_before_NIR)
```

# Image After RGB & FCC

```python
img_after_R = img_after[:, :, 2]
img_after_G = img_after[:, :, 1]
img_after_B = img_after[:, :, 0]
img_after_N = img_after[:, :, 3]
```

```python
img_after_RGB = np.stack([img_after_R, img_after_G, img_after_B], axis=2)
```

```python
io.imshow(img_after_RGB)
```

```python
# io.imsave("outputs/img_after_RGB.png", img_after_RGB)
```

```python
img_after_NIR = np.stack([img_after_R, img_after_G, img_after_N], axis=2)
```

```python
io.imshow(img_after_N)
```

```python
# io.imsave("outputs/img_after_FCC.png", img_after_NIR)
```

```python
diff = abs(img_before_RGB - img_after_RGB)
diff = exposure.rescale_intensity(diff)
```

```python
io.imshow(diff)
```
